```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python

x = data.drop(columns=['Customer ID', 'Subscription Status'])
y=data["Subscription Status"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_features=["Gender","Item Purchased","Category","Location","Size","Color","Season","Payment Method","Shipping Type","Discount Applied","Promo Code Used","Frequency of Purchases"]
```

```python
x.info()
```

```python
numeric_features = x.select_dtypes(include=['int64', 'float64']).columns
categorical_features = x.select_dtypes(include=['object']).columns

# One-hot encode the categorical features
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)], remainder='passthrough')
transformed_x = transformer.fit_transform(x)
```

```python
pd.DataFrame(transformed_x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.2, random_state=42)

# Scale numerical features
scaler = StandardScaler(with_mean=False)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_preds)
conf_matrix = confusion_matrix(y_test, y_preds)
class_report = classification_report(y_test, y_preds)

print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
from sklearn.model_selection import cross_val_score


cv_scores = cross_val_score(model, transformed_x, y, cv=10)

print("Cross-validation scores:", cv_scores)
```

```python
print("Mean CV accuracy: {:.2f}%".format(cv_scores.mean()*100))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model=LogisticRegression(random_state=42)

# Train the model
logistic_model.fit(x_train, y_train)
logistic_model.score(x_test,y_test)
```

```python
logistic_cv=cross_val_score(logistic_model,transformed_x,y,cv=10)
print("the logistic cv is",round(logistic_cv.mean()*100,2),"%")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming you have trained your Random Forest Classifier and made predictions (y_pred)

# Compute precision, recall, and F1-score
precision = precision_score(y_test, y_preds, pos_label='Yes')
recall = recall_score(y_test, y_preds, pos_label='Yes')
f1 = f1_score(y_test, y_preds, pos_label='Yes')

# Print the metrics
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
